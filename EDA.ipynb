#Hypothesis (H0): There is significant difference between the two target groups w.r.t the age of the building

#First we'll check whether the TARGET columns are normally distributed or not. We'll use the Anderson Darling test, which is less sensitive to outliers

```python
for j in ['YEARS_BEGINEXPLUATATION_AVG','YEARS_BEGINEXPLUATATION_MEDI','YEARS_BEGINEXPLUATATION_MODE']:
  df[j].fillna(df[j].mode()[0],inplace=True)
```

```python
from scipy.stats import anderson
for j in ['YEARS_BEGINEXPLUATATION_AVG','YEARS_BEGINEXPLUATATION_MEDI','YEARS_BEGINEXPLUATATION_MODE']:
  result = anderson(df[j])
  print('Normal Distribution check for column ',j)
  print('Statistic:', result.statistic)
  if result.statistic < result.critical_values[2]:  # Compare with 5% significance level
      print("Sample appears to be normally distributed (fail to reject H0)")
  else:
      print("Sample does not appear to be normally distributed (reject H0)")
  print('\n')
```

#**Here we perform Hypothesis Testing**

```python
import scipy.stats as stats
for j in ['YEARS_BEGINEXPLUATATION_AVG','YEARS_BEGINEXPLUATATION_MEDI','YEARS_BEGINEXPLUATATION_MODE']:
  print('checking for column ',j)
  a1=df[df['TARGET']==0][j]
  a2=df[df['TARGET']==1][j]
  u_statistic, p_value = stats.mannwhitneyu(a1,a2)
  print(p_value)
  if p_value<0.05:
    print('Reject H0')
  else:
    print('Fail to reject H0')
  print('\n')
```

#Hypothesis 1:
#H0: Owning a car and loan repayment are independent of each other
#We'll use the chi square test since the car column is categorical  

```python
chi_square_df = pd.crosstab(df['FLAG_OWN_CAR'], df['TARGET'])

chi2, p_value, dof, expected = stats.chi2_contingency(chi_square_df)
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p_value}")
if p_value<0.05:
    print('Reject H0')
else:
    print('Fail to reject H0')
```
